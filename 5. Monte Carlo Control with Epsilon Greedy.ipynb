```python
# Import Packges and Enviroments
import torch
import gym
from collections import defaultdict
env=gym.make('Blackjack-v0')
```

# STEP 1: Develop Run Episode Function with Greedy

```python
def run_episode(env, Q, epsilon, n_action):
    #1.Same initial variables as always, capturing the inital state and lists to record the rewards+1, actions and states
    state=env.reset()
    rewards=[]
    actions=[]
    states=[]
    is_done=False
    #2.Initialize the while loop that breaks whe the condition is satisfied
    while not is_done:
    #3.This creates a tensor(1,1)*epsilon/2 , 0<epsilon<1 , if epsilon is 1 all actions(0,1) taken with probability of one
        probs=torch.ones(n_action)*epsilon/n_action
    #4.Lets remeber that Q[state] is a tensor[xxx,xxx] with two posible actions in the case of blackjack so the result here is either 0, 1
        best_action=torch.argmax(Q[state]).item()
        probs[best_action]+=1-epsilon
        action=torch.multinomial(probs,1).item()
        actions.append(action)
        states.append(state)
        state, reward, is_done, infor=env.step(action)
        rewards,append(reward)
        if is_done:
            break
    return states, actions, rewards
```
